```python
def save_historical(symbol):
    filesymbol = [p[0] for p in matches if p[1]==symbol]
    filesymbol = filesymbol[0] if filesymbol else None
    if filesymbol:
        file = glob.glob(f'../data/historical futures/{filesymbol}_*')[0]
        df = pd.read_csv(file,
                         sep=' ',
                         skiprows=[0],
                         names=['date','open','high','low','close','oi','volume'],
                         parse_dates=['date'])\
                        .drop(labels=['oi','volume'], axis=1)\
                        .dropna()\
                        .round(products[symbol].decimal_len)
        #df.columns = ['date','open','high','low','close','oi','volume']
        #df.set_index('date', inplace=True)
        df['date'] = df['date'].astype('int64')
        db.create_dataset(symbol, data=df.values)
        print(f'{symbol} saved')
```

```python
for p in products.keys():
    save_historical(p)
```

```python
def historical_ohlc(symbol):
    db = h5py.File('../data/historical_ohlc.h5py', 'r')
    if symbol in db.keys():
        data = db[symbol][:]
        df = pd.DataFrame(data[:,[1,2,3,4]], index=data[:,0].astype('M8[ns]'), columns=['open','high','low','close'])
        db.close()
        return df
    else:
        print(f"There is no historical data for {symbol}")
```
